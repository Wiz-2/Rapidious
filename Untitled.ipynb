```python
#filling the null data with numeric mean
df_no_duplicates_filled = df_no_duplicates.fillna(df.mean(numeric_only = True))
```

```python
df_no_duplicates_filled.shape
```

```python
df_no_duplicates_filled.describe()
```

```python
# Convert 'quick & easy' and 'quick and healthy' to boolean
df_no_duplicates_filled['quick_easy_bool'] = df_no_duplicates_filled['quick & easy'].astype(bool)
df_no_duplicates_filled['quick_healthy_bool'] = df_no_duplicates_filled['quick and healthy'].astype(bool)

# Create 'quick_both' as logical AND of 'quick_easy_bool' and 'quick_healthy_bool'
df_no_duplicates_filled['quick_both'] = df_no_duplicates_filled['quick_easy_bool'] & df_no_duplicates_filled['quick_healthy_bool']
df_no_duplicates_filled['quick_both'] = df_no_duplicates_filled['quick_both'].astype(int)

# Create 'neither_quick_either' as logical NOT of both 'quick_easy_bool' and 'quick_healthy_bool'
df_no_duplicates_filled['neither_quick_either'] = (~df_no_duplicates_filled['quick_easy_bool']) & (~df_no_duplicates_filled['quick_healthy_bool'])
df_no_duplicates_filled['neither_quick_either'] = df_no_duplicates_filled['neither_quick_either'].astype(int)

# Calculate average ratings for each category
# Quick & Easy vs Others
avg_rating_quick_easy = df_no_duplicates_filled.groupby('quick & easy')['rating'].mean().reset_index()
avg_rating_quick_easy.columns = ['Quick & Easy', 'Average Rating']

# Quick & Healthy vs Others
avg_rating_quick_healthy = df_no_duplicates_filled.groupby('quick and healthy')['rating'].mean().reset_index()
avg_rating_quick_healthy.columns = ['Quick & Healthy', 'Average Rating']

# Quick & Both vs Others
avg_rating_quick_both = df_no_duplicates_filled.groupby('quick_both')['rating'].mean().reset_index()
avg_rating_quick_both.columns = ['Quick & Both', 'Average Rating']

# Neither Quick & Easy nor Quick & Healthy vs Others
avg_rating_neither = df_no_duplicates_filled.groupby('neither_quick_either')['rating'].mean().reset_index()
avg_rating_neither.columns = ['Neither Quick & Easy nor Quick & Healthy', 'Average Rating']

# Combine average ratings into one DataFrame for plotting
avg_ratings = pd.DataFrame({
    'Category': [
        'Quick & Easy', 
        'Quick & Healthy', 
        'Quick & Both', 
        'Neither'
    ],
    'Average Rating': [
        avg_rating_quick_easy.loc[avg_rating_quick_easy['Quick & Easy'] == 1, 'Average Rating'].values[0],
        avg_rating_quick_healthy.loc[avg_rating_quick_healthy['Quick & Healthy'] == 1, 'Average Rating'].values[0],
        avg_rating_quick_both.loc[avg_rating_quick_both['Quick & Both'] == 1, 'Average Rating'].values[0],
        avg_rating_neither.loc[avg_rating_neither['Neither Quick & Easy nor Quick & Healthy'] == 1, 'Average Rating'].values[0]
    ]
})

# Plotting the Average Ratings by Category
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Average Rating', data=avg_ratings, palette='Set2')

# Add annotations for average ratings
for index, row in avg_ratings.iterrows():
    plt.text(index, row['Average Rating'] + 0.02, f"{row['Average Rating']:.2f}", color='black', ha="center")

plt.title('Average Recipe Ratings by Quick Categories')
plt.xlabel('Category')
plt.ylabel('Average Rating')
plt.ylim(3, 5)  # Adjust based on your rating scale
plt.tight_layout()
plt.show()
```

# Business Insight:
Having the highest mean rating (3.83) implies that users tend to prefer easy and quick recipes the most. 
Quick & Healthy recipes earn the lowest mean rating (3.21) signaling that the users might be disinterested or find these recipes difficult.
Recipes that are neither quick nor healthy also receive relatively high ratings (3.67) meaning that users appreciate quality or richness in the non-quick recipes.

# Actionable Strategy:
Focus on promoting Quick & Easy recipes in your marketing efforts as they are the most palatable to users.
Research Quick & Healthy recipes in a bid to make them more interesting, perhaps by minimizing the ingredients or the preparation.
Keep on providing non-quick but elaborate recipes for users who may want something richer or more labor-intensive.

```python
# List of tags (ensure these match exactly with your DataFrame's column names)
tags = ['#cakeweek', '#wasteless', '22-minute meals', '3-ingredient recipes', 
        '30 days of groceries', 'advance prep required', 'back to school', 
        'backyard bbq', 'bastille day', 'bon appétit', 'birthday', 'christmas', 
        'christmas eve', 'cinco de mayo', 'cocktail party', 'diwali', 
        'dorie greenspan', 'epi + ushg', 'epi loves the microwave', 
        'fall', 'family reunion', "father's day", 'flaming hot summer', 
        'fourth of july', 'friendsgiving', 'graduation', 'halloween', 
        'hanukkah', 'house & garden', 'house cocktail', "mother's day", 
        "new year's day", "new year's eve", 'no meat, no problem', 
        'no sugar added', 'quick & easy', 'quick and healthy', 
        'ramadan', 'rosh hashanah/yom kippur', 'sukkot', 
        'super bowl', "st. patrick's day", 'tailgating', 
        'thanksgiving', "valentine's day", 'lunar new year', 
        'tested & improved']

# Ensure all tags are present in the DataFrame
valid_tags = [tag for tag in tags if tag in df_no_duplicates_filled.columns]

# Step 1: Reshape the DataFrame from wide to long format
melted_df = df_no_duplicates_filled.melt(
    id_vars=['rating'], 
    value_vars=valid_tags,
    var_name='Tag', 
    value_name='Presence'
)

# Step 2: Filter to include only rows where the tag is present (Presence == 1)
melted_df = melted_df[melted_df['Presence'] == 1]

# Step 3: Group by 'Tag' to calculate count of recipes (for color mapping)
grouped = melted_df.groupby('Tag').agg(
    Number_of_Recipes=('rating', 'count')
).reset_index()

# Step 4: Normalize the number of recipes for color mapping
norm = plt.Normalize(grouped['Number_of_Recipes'].min(), grouped['Number_of_Recipes'].max())
cmap = plt.cm.viridis  # Choose a colormap (e.g., 'viridis', 'plasma', 'inferno')

# Step 5: Merge the normalized colors into the original DataFrame
melted_df = pd.merge(melted_df, grouped, on='Tag')

# Step 6: Create the box plot
plt.figure(figsize=(20, 10))  # Adjust the figure size as needed

# Loop through tags and plot each boxplot with its respective color based on the number of recipes
for i, tag in enumerate(grouped['Tag']):
    sns.boxplot(
        x='Tag', y='rating', data=melted_df[melted_df['Tag'] == tag], 
        color=cmap(norm(melted_df[melted_df['Tag'] == tag]['Number_of_Recipes'].values[0])),
        fliersize=0
    )

# Add a color bar to indicate the number of recipes
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])  # Only needed for older Matplotlib versions
plt.colorbar(sm, label='Number of Recipes')

# Set plot titles and labels
plt.title('Recipe Ratings by Tag (Box Plot) with Color Indicating Number of Recipes', fontsize=16)
plt.xlabel('Tags', fontsize=14)
plt.ylabel('Rating', fontsize=14)
plt.xticks(rotation=90, fontsize=10)  # Rotate x-axis labels for readability

# Adjust layout to prevent clipping of tick-labels
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# List of meal columns (ensure these match your DataFrame)
meal_columns = ['breakfast', 'lunch', 'dinner', 'snack','brunch']  # Replace with actual meal columns

# Step 1: Calculate the sum of recipes per meal type
meal_counts = df_no_duplicates_filled[meal_columns].sum()

# Step 2: Calculate the average rating for each meal type
# Assuming ratings are in a 'rating' column and 1/0 binary presence columns for meals
average_ratings = df_no_duplicates_filled.groupby(meal_columns)['rating'].mean()

# Step 3: Normalize the average ratings for color mapping
norm = plt.Normalize(average_ratings.min(), average_ratings.max())
cmap = plt.cm.viridis  # Choose colormap

# Generate colors based on normalized average ratings
colors = [cmap(norm(avg_rating)) for avg_rating in average_ratings]

# Step 4: Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(meal_counts, labels=meal_columns, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Distribution of Recipes by Meal Type with Color Indicating Average Rating')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular

# Add a color bar to indicate the mapping of colors to average ratings
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])  # Only needed for older Matplotlib versions
plt.colorbar(sm, label='Average Rating')

# Display the plot
plt.show()
```
